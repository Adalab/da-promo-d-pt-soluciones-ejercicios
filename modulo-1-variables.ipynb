```python
# Crear las variables iniciales
gatos = 8  # Variable tipo int
perros = "cuatro"  # Variable tipo str de seis caracteres

# Imprimir las variables
print(gatos)
print(perros)

# Concatenar y mostrar texto
print(perros + " más " + perros)

# Construir una oración usando variables
print(f"{perros} más {perros} es igual a {gatos}")
print(f"{perros} + {perros} = {gatos}")

# Definir variables con diferentes tipos
pizzas = 2.0  # Float
magdalenas = "2"  # String
postres = 0  # Integer

# Comprobar tipos de datos
print(type(pizzas))  # Tipo de pizzas
print(type(magdalenas))  # ¿Es string magdalenas?

# Redefinir postres
postres = 0

# Operaciones aritméticas
print(gatos + pizzas)  # Suma
print(gatos - pizzas)  # Diferencia
print(gatos * 2)  # Multiplicación

# Calcular piezas
piezas = 8 * pizzas
piezas += 3

# Calcular el cuadrado de un número
superficie = 9 ** 2

# Operaciones entre variables
print(superficie / gatos)  # Superficie por gato
print(piezas // gatos)  # Piezas enteras por gato
print(piezas % (gatos * 21))  # Piezas restantes

# Redondear dentro del print
print(round(superficie / gatos, 1))

# Comparaciones lógicas
print(gatos > pizzas)
print(gatos < superficie)
print(2 * pizzas != gatos)
print(not (2 * pizzas == gatos))
print(piezas >= pizzas and superficie != 0)
print(piezas >= pizzas and piezas < superficie)

# Comparación avanzada
print((pizzas > gatos) or (piezas % gatos == 0))

# Sin usar la variable piezas más de una vez
print((8 * pizzas + 3) % gatos == 0)

# Comparación diferente de cero sin usar !=
print(bool(superficie))

# Uso de input() para preguntar fechas
dia_cumple = input("Introduce tu día de cumpleaños: ").strip()
mes_cumple = input("Introduce tu mes de cumpleaños: ").strip()

# Formar la fecha completa
fecha_cumple = dia_cumple + " " + mes_cumple
print("Fecha de cumpleaños es " + fecha_cumple)

# Mostrar la variable perros en mayúsculas
print(perros.upper())
```
