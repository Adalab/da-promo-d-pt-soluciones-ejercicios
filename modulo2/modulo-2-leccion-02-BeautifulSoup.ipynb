```python
# como siempre...primero las librerias!

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Definir la URL de la página web
url_medallero = "https://es.wikipedia.org/wiki/Anexo:Medallero_de_los_Juegos_Ol%C3%ADmpicos_de_Tokio_2020"

# Realizar una solicitud GET a la URL
res_medallero = requests.get(url_medallero)

# Verificar el código de estado de la respuesta
print("Código de estado:", res_medallero.status_code)
```

```python
# Crear un objeto BeautifulSoup para analizar el contenido HTML
sopa_medallero = BeautifulSoup(res_medallero.content, "html.parser")

# Encontrar la tabla que contiene los datos del medallero
tabla = sopa_medallero.find("table")

# Encontrar todas las filas de la tabla
filas = tabla.find_all("tr")
```

```python
# podemos averiguar cuantas tablas tiene mi pagina a investigar!
tablas = sopa_medallero.find_all("table")
len(tablas)
```

Si bien hay 3 tablas, a nosotras solo nos insteresa la primera.

```python
# Inicializar una lista para almacenar los resultados de mis filas
resultados = []

# Iterar a través de las filas (ignorando la primera y la última)
for fila in filas[1:-1]:
    # Imprimir el contenido de la fila (puede comentarse o eliminarse)
    print(fila.text)
    
    # Dividir el contenido de la fila en columnas
    dividir_fila = fila.text.split("\n")[1:-1]
    
    # Intentar convertir los elementos en números enteros si es posible, de lo contrario, mantenerlos como texto
    for indice, elemento in enumerate(dividir_fila):
        try:
            dividir_fila[indice] = int(elemento)
        except:
            dividir_fila[indice] = elemento.strip()
    
    # Agregar la fila procesada a la lista de resultados
    resultados.append(tuple(dividir_fila))
```

```python
# buscando los encabezados de mi tabla...

lista_encabezados = tabla.find_all("th")

print("La lista que nos devuelve el metodo '.find_all()' es:\n", lista_encabezados)

# como hemos estado haciendo hasta ahora, tendremos que iterar por la lista obtenida en el paso anterior y extraer el texto de cada elemento
encabezados = [columna.text for columna in lista_encabezados]

print("\n-------------------------\n")
print("Los encabezados de la tabla son:", encabezados)
```

Parece ser que algunos nombres de mi encabezado estan mal...se ven raros!!! probablemente es una imagen, pero hasta las imagenes tienen descripcion...lo veremos mas adelante

```python
# convertir nuestros resultados en un dataframe

df_medallas = pd.DataFrame(resultados)
df_medallas
```

```python
# agregando el encabezado a mi dataframe usando .columns aplicado al dataframe que quiero usar.

df_medallas.columns = encabezados
df_medallas
```

```python
# usamos metodos de pandas para dejar nuestra tabla lo mas prolija posible y eliminamos la columna de posicion cero, en el eje 1.

df_medallas = df_medallas.drop(df_medallas.columns[0], axis=1)
df_medallas
```

Vemos que las cabeceras no son todas del tipo texto...podriamos por ejemplo elegir la posicion 2 de nuestra lista y utilizar el metodo find para buscar "img" ya que vimnos en la lista que ahi dentro hay una palabra... 

```python
img_tag = nueva_lista[2].find("img")

# Extraer el valor del atributo 'alt'
alt_text = img_tag['alt']

print(alt_text) 
```

Habiendo dicho esto, podriamos hacer un ciclo for para tratar a estos elementos particulares y aquellos que simplemente sean un texto podriamos "cortarlos" para solo quedarnos con la palabra.

```python
# instaciamoa la variable "nueva_lista" y buscamos todas las cabeceras ("th")
nueva_lista = tabla.find_all("th")

#creamos la lsita vacia donde almacenaremos las palabras de cabecera
headers = []

# ciclo for para tratar segun corresponda a los elementos de mi lista
for th in nueva_lista:
    img = th.find("img")
    if img and img.get("alt"):
        text = img.get("alt")
    else:
        text = th.get_text(strip=True)
    headers.append(text)
```
