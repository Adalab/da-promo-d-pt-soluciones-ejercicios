3. Crea un *array* unidimensional con 50 elementos con el método que prefieras y extrae: 

    - El primer elemento.


    - Lo elementos que están en posición impar.


    - Los cuatro primeros elementos.


    - Los últimos tres elementos.


    - Los elementos entre el 4 y el 8 (incluido). 

4. Crea un *array* de dos dimensiones de 4 x 5 con el método que prefieras y extrae:

    - La primera fila y todas las columnas.


    - Todas las filas pares.


    - La primera columna y todas las filas.


````python
"""
1. Crea un array unidimensional con 50 elementos con el método que prefieras y extrae: 

    - El primer elemento.


    - Lo elementos que están en posición impar.


    - Los cuatro primeros elementos.


    - Los últimos tres elementos.


    - Los elementos entre el 4 y el 8 (incluido). 

"""

# Crear un array unidimensional con 50 elementos
array_50 = np.arange(1, 51)

# Extraer el primer elemento
primer_elemento = array_50[0]

# Extraer los elementos en posiciones impares
elementos_impares = array_50[1::2]

##
## 1 indica el índice de inicio (comienza desde el segundo elemento, ya que los índices en Python empiezan en 0).
## :: es el separador de los parámetros de slicing.
## 2 es el paso, lo que significa que se selecciona cada segundo elemento a partir del índice de inicio.
## En resumen, array_50[1::2] selecciona todos los elementos de array_50 que están en posiciones impares (segundo, cuarto, sexto, etc.)

# Extraer los cuatro primeros elementos
primeros_cuatro = array_50[:4]

# Extraer los últimos tres elementos
ultimos_tres = array_50[-3:]

# Extraer los elementos entre el 4 y el 8 (incluidos)
entre_4_y_8 = array_50[4:9]

# Imprimir los resultados
print("Primer elemento:", primer_elemento)
print("Elementos en posiciones impares:", elementos_impares)
print("Cuatro primeros elementos:", primeros_cuatro)
print("Últimos tres elementos:", ultimos_tres)
print("Elementos entre el 4 y el 8 (incluidos):", entre_4_y_8)
````

````python
"""
    2. Crea un array de dos dimensiones de 4 x 5 con el método que prefieras y extrae:

    - La primera fila y todas las columnas.


    - Todas las filas pares.


    - La primera columna y todas las filas.
"""

# Crear un array bidimensional de 4x5 usando el metodo random.randint()
array_2d = np.random.randint(0, 50, (4,5))

# Extraer la primera fila y todas las columnas
primera_fila_todas_columnas = array_2d[0, :]

# Extraer todas las filas pares
todas_filas_pares = array_2d[::2, :]

# Extraer la primera columna y todas las filas
primera_columna_todas_filas = array_2d[:, 0]

# Imprimir los resultados
print("mi array 2D es:",array_2d)

print("\nPrimera fila y todas las columnas:",primera_fila_todas_columnas)

print("\nTodas las filas pares:", todas_filas_pares)

print("\nPrimera columna y todas las filas:",primera_columna_todas_filas)
````

    - Crea una mascara o filtro donde los valores mayores a 30 y menores a 10 se reemplace por "objetivo" y el resto por "no es mi objetivo".


    - Los valores de mi array bidimensional multiplicalos por el escalar 10.


6. Crea un nuevo *array* de dos dimensiones con la misma forma que el anterior y con el método que prefieras y realiz lo siguiente:

    - Suma el primer array con el segundo.


    - Aplicar la potencia para el primer array elevado al segundo.

````python
"""
1. Crea un *array* bidimensional con 20 elementos aleatrorios entre 0 y 50 con el método que prefieras y: 

    - Crea una mascara o filtro donde los valores mayores a 30 y menores a 10 se reemplace por "objetivo" y el resto por "no es mi objetivo".


    - Los valores de mi array bidimensional multiplicalos por el escalar 10.
"""

# creando mi array de 20 elementos aleatorios entre 0 y 50
array = np.random.randint(0, 51, size=(4, 5))
print("mi array")
print(array)

# creando la mascara
mascara = np.where((array > 30) | (array < 10), "objetivo", "no es mi objetivo")
print("\nMáscara:")
print(mascara)

# Multiplicar el array original por el escalar 10
array_mult = array * 10
print("\nArray multiplicado por 10:")
print(array_mult)
````

````python
"""
2. Crea un nuevo *array* de dos dimensiones con la misma forma que el anterior y con el método que prefieras y realiza lo siguiente:

    - Suma el primer array con el segundo.


    - Aplicar la potencia para el primer array elevado al segundo.
"""

# Crear un nuevo array bidimensional con la misma forma
nuevo_array = np.random.randint(0, 51, size=(4, 5))
print("Nuevo array:\n", nuevo_array)

print("-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-")

# mostrar el primer array
print("Primer array:\n", array)

print("-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-")

# Sumar el primer array con el segundo
suma_arrays = array + nuevo_array
print("Suma de los arrays:\n", suma_arrays)

print("-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-")

# Aplicar la potencia del primer array elevado al segundo
potencia_arrays = np.power(array, nuevo_array) # otra forma seria: potencia_arrayas = array ** nuevo_array
print("Potencia del primer array elevado al segundo:\n", potencia_arrays)
````

7. Crea un *array*, puede ser bidimensional! Eso si, los valores tienen que estar entre 0-100

    - Calcula: 

        - La media de los elementos
        - La media de las columnas y redondéalo a dos decimales
        - El valor máximo de cada fila
        - El valor mínimo de cada fila
        - Hay algun valor que sea mayor que 90 en tu *array* 
        - Donde los podemos encontrar, cuáles son sus índices
        - Ordena el *array* de mayor a menor

    📝 **NOTA** En este ejercicio os hemos pedido algunos de los métodos, pero sentiros libres de practicar otros para ir asentando los conocimientos. 

````python
"""
    1. Crea un *array*, puede ser bidimensional! Eso si, los valores tienen que estar entre 0-100

    - Calcula: 

        - La media de los elementos
        - La media de las columnas y redondéalo a dos decimales
        - El valor máximo de cada fila
        - El valor mínimo de cada fila
        - Hay algun valor que sea mayor que 90 en tu *array* 
        - Donde los podemos encontrar, cuáles son sus índices
        - Ordena el *array* de mayor a menor
"""

# Crear un array bidimensional de 5x5 con valores aleatorios entre 0 y 100
array = np.random.randint(0, 101, size=(5, 5))

print("Array original:")
print(array)
print()
print("-----------------------------------------------")
# Calcular la media de todos los elementos del array
media_total = np.mean(array)
print("Media de todos los elementos:", media_total)

print("-----------------------------------------------")
# Calcular la media de cada columna y redondear a dos decimales
media_columnas = np.round(np.mean(array, axis=0), 2)
print("Media de cada columna:")
print(media_columnas)

print("-----------------------------------------------")
# Calcular el valor máximo de cada fila
maximos_filas = np.max(array, axis=1)
print("Valor máximo de cada fila:")
print(maximos_filas)

print("-----------------------------------------------")
# Calcular el valor mínimo de cada fila
minimos_filas = np.min(array, axis=1)
print("Valor mínimo de cada fila:")
print(minimos_filas)

print("-----------------------------------------------")
# Verificar si hay algún valor mayor que 90 en el array
hay_mayor_90 = np.any(array > 90)
print("¿Hay algún valor mayor que 90 en el array?", hay_mayor_90)

# Encontrar los índices de los valores mayores que 90
if hay_mayor_90:
    indices_mayor_90 = np.where(array > 90)
    print("Índices de los valores mayores que 90:")
    print(indices_mayor_90)

print("-----------------------------------------------")
# Ordenar el array de mayor a menor
array_ordenado = np.sort(array.flatten())[::-1] # con el metodo flatten aplanamos el array de dos dimensiones a una sola dimension
print("Array ordenado de mayor a menor:")
print(array_ordenado)

print("-----------------------------------------------")
````
