1- Abre el fichero de `rating-and-performance` con el método que sea más correcto. Además muestra las 5 primeras filas del *DataFrame*. 

```python
# Paso 1: Abrir el archivo y mostrar las 5 primeras filas del DataFrame
df = pd.read_csv('rating-and-performance.csv') 
# si esta metido en alguna carpeta por ejemplo que se llame files, tendriamos que 
# acceder a la carpeta y despues al archivo --> ('files/rating-and-performance.csv')

# Eliminar la primera columna (seleccionando todo menos la primera columna)
df = df.iloc[:, 1:]
print("Primeras 5 filas del DataFrame:")
df.head()
```

2- Pongamos en práctica el `loc` e `iloc`:

```python
# Paso 2: Usando loc e iloc

# Usando loc:

# Extraer todas las filas de la columna "price"
prices_loc = df.loc[:, 'price']
print("\ndatos de las filas de 'price' usando loc:\n", prices_loc)
print("\nTipo de datos de la columna 'price' usando loc:", prices_loc.dtype)
```

```python
# Extraer todas las filas de las columnas "price" y "retail_price"
# ¿De qué tipo es el resultado obtenido?
prices_retail_loc = df.loc[:, ['price', 'retail_price']]
print("\ndatos de las filas de 'price' y 'retail_price' usando loc:\n", prices_retail_loc)
print("-----------------------------------------------------------")
print("\nTipo de datos de la columna 'price' y 'retail_price'usando loc:", prices_retail_loc.dtypes)
```

```python
# Extraer las filas pares de las columnas "rating", "rating_count", "rating_five_count", "rating_four_count", 
# "rating_three_count", "rating_two_count", "rating_one_count"
filas_pares_loc = df.loc[::2, ['rating', 'rating_count', 'rating_five_count', 'rating_four_count', 'rating_three_count', 'rating_two_count', 'rating_one_count']]
filas_pares_loc.head()
```

```python
# Extraer las filas cuyo precio sea mayor que 10 y sus unidades vendidas sea mayor que 5000
filtro_df_loc = df.loc[(df['price'] > 10) & (df['units_sold'] > 5000)]
filtro_df_loc.head()
```

```python
# Usando iloc (obtendremos los mismos resultados que con loc):

# Extraer todas las filas de la columna "price"
prices_iloc = df.iloc[:, 2]
prices_iloc
```

```python
# Extraer todas las filas de las columnas "price" y "retail_price"
prices_retail_iloc = df.iloc[:, [2, 3]]
prices_retail_iloc
```

```python
# Extraer las filas pares de las columnas "rating", "rating_count", "rating_five_count", "rating_four_count", 
# "rating_three_count", "rating_two_count", "rating_one_count"
filas_pares_iloc = df.iloc[::2, 7:14]
filas_pares_iloc.head()
```

4. Crea una serie con nuemeros correlativos entre 1 y 12 inclusive (puedes hacerlo como lista o diccionario): 

    - Cambia el indice de la serie anteriormente creada para que comience con el numero 20.


    - Nuevamente cambia el indice de la serie, esta vez con una lista que haga referencia a los meses del año (tenemos 12 elementos como la cantidad de meses del año).

```python
# Paso 1: Crear una serie con números correlativos entre 1 y 12 inclusive
serie = pd.Series(range(1, 13))
print("Serie inicial:\n", serie)
```

```python
# Paso 2: Cambiar el índice de la serie para que comience con el número 20
serie.index = range(20, 32)
print("\nSerie con índice comenzando en 20:\n", serie)
```

```python
# Paso 3: Cambiar el índice de la serie con una lista que haga referencia a los meses del año
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 
         'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
serie.index = meses
print("\nSerie con índice de meses del año:\n", serie)
```

```python
# Paso 4: Mostrar los valores que están en la posición 3 y 8 de la serie
valor_posicion_3 = serie.iloc[3]  # Recordar que iloc usa índice basado en 0
valor_posicion_8 = serie.iloc[8]
print("\nValor en la posición 3:", valor_posicion_3)
print("Valor en la posición 8:", valor_posicion_8)
```

```python
# Paso 5: Mostrar los valores que están etiquetados en enero y agosto
valor_enero = serie['Enero']
valor_agosto = serie['Agosto']
print("\nValor en Enero:", valor_enero)
print("Valor en Agosto:", valor_agosto)
```

```python
# Paso 6: Mostrar valores, índices, forma, tamaño y el tipo de elementos de la serie
print("\nValores de la serie:\n", serie.values)
print("Índices de la serie:\n", serie.index)
print("Forma de la serie:", serie.shape)
print("Tamaño de la serie:", serie.size)
print("Tipo de elementos de la serie:", serie.dtype)
```
