```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df_insight = pd.read_csv("insight.csv")
df_insight.head(2)
```

```python
df_rating = pd.read_csv("rating-and-performance.csv")
df_rating.head(2)
```

```python
# 1. Unir los dos DataFrames en uno solo, usanado el método el método que considereis más correcto. 
df = df_rating.merge(df_insight, left_on = "merchant_id", right_on = "merchant.id", how = "inner", suffixes = ["_sold", "_employee"] )
df.head(2)
```

```python
# 2. Explorad el nombre de las columnas, ¿hay alguna que este repetida o que no nos aporte información de valor? 
# ¿Hay alguna columna que solo tenga un valor? En caso de ser así, eliminadlas. 
# **NOTA** Hay tres columnas que cuando exploramos sus valores únicos y sus frecuencias ayer, nos deberíamos haber dado cuenta que solo 
# tiene un valor.  
```

```python
df.columns
```

```python
# las columnas que solo tienen un valor y que por tanto eliminaremos son: 
df.columns[df.nunique() == 1]
```

```python
# las columnas index_sold y index_employee no aportan info de valor, las genera por defecto pandas al cargar los dos csv

# las columnas title, y title_orig tienen informacion muy similar, eliminamos title_orig, lo mismo pasa con merchant_id y merchant.id, 
# por lo que eliminamos una de ellas
```

```python
df.drop(["index_sold", "index_employee", "title_orig", "merchant.id", 'currency_buyer', 'has_urgency_banner', 'theme', 'crawl_month'], axis = 1, inplace = True)
df.head(2)
```

```python
# 3- Algunas de las columnas tienen `.` en sus nombres y en otras tenemos `_`.  Unifica el nombre de las columnas para que todas tengan "_". 


# Forma 1 de resolverlo: con dictionary comprehension:
nuevas_columnas = {col: col.replace(".", "_").lower() for col in df.columns}

df.rename(columns = nuevas_columnas, inplace = True)
df.head()
```

```python
# Otra forma de resolverlo, con un bucle for 

diccionario_nuevas_columnas =  {}

for col in df.columns:
    diccionario_nuevas_columnas[col] = col.replace('.', '_').lower()
    # creamos el diccionario en cada iteracion del bucle for, para cada columna se creara una clave-valor, donde la clave
    # será el nombre antiguo de la columna (col) y el valor será el nuevo valor despues de aplicar replace y lower.

df.rename(columns=diccionario_nuevas_columnas, inplace=True)

df.columns
```

```python
#  Usando el método de limpieza, elimina de la columna "merchant_info_subtitle" la parte del *string* de los paréntesis. 

# con regex
df["merchant_info_subtitle"]  = df["merchant_info_subtitle"].str.replace(r"\(.*\)", "", regex = True)
# con regex sustituimos todo lo que está dentro del parentesis por un string vacio "" usando replace

# sin regex
df['merchant_info_subtitle'] = df['merchant_info_subtitle'].str.split('(', expand=True).get([0])
# al hacer split por el parentesis se nos genera una lista con dos elementos, lo que hay antes del parentesis (elemento 0 que es 
# el que nos interesa) y el elemento uno que tendra la info que había justo despues de (, que es lo que hemos usado como punto de corte

df.head()
```

```python
# Si nos fijamos, en la primera fila antes solo teníamos un paréntesis, pero los hemos eliminado en el paso anterior, 
# y ahora tenemos un string vacío, lo cual no tiene mucho sentido. 
# En este ejercicio debereis reemplazar esos strings vacios por nulos de Pandas (NOTA estos deben ser indicados de la siguiente forma np.nan, 
# y deberéis importar numpy).

df["merchant_info_subtitle"]  = df["merchant_info_subtitle"].replace("", np.nan)
df.head(1)
```

```python
# Hay ciertas columnas que no vamos a necesitar para nuestro análisis. En concreto, vamos a eliminar las columnas de badges_count, 
# product_variation_inventory, urgency_text, merchant_title, merchant_has_profile_picture, merchant_profile_picture, product_picture, 
# total_urgency_count, has_urgency_banner y urgency_text_rate. Guardad los cambios realizados hasta este punto en un csv para usarlo en 
# los ejercicios de la lección de mañana.

# has_urgency_banner ya la eliminamos en el ejercicio 2, no la añado en la lista de columnas a eliminar porque daría error.
eliminar = ["badges_count", "product_variation_inventory","urgency_text", 'merchant_title', 'merchant_has_profile_picture',  'merchant_profile_picture', 'product_picture', 'total_urgency_count', 'urgency_text_rate']
df.drop(eliminar, axis = 1, inplace = True)
# tambien podríamos poner: df.drop(columns=eliminar, inplace=True)
```

### 7.Pongamos en práctica el filtrado de datos en Pandas:



```python
# Seleccionad aquellos productos cuyo color es verde. ¿Cuántos productos tenemos?
df_verde  = df[df["product_color"] == "green"]
df_verde.shape
```

```python
# Seleccionad aquellos productos cuyo color es verde y rojo. ¿Cuántos productos tenemos de cada tipo de color?
colores = ["red", "green"]

df_colores = df[df["product_color"].isin(colores)]

# otra forma de hacerlo (aunque es mas comodo con isin) sería:
# df_colores = df[(df['product_color'] == 'red')&(df['product_color'] == 'green')]

df_colores["product_color"].value_counts()
```

```python
# Seleccionar aquellos productos de la talla M, que valgan más de 15. ¿Cuántos productos tenemos?
condicion1 = df["product_variation_size_id"] == "M"
condicion2 = df["price"] > 15

df_talla_precio = df[condicion1 & condicion2]
df_talla_precio.shape
```

```python
# Seleccionad aquellos productos cuyo precio este entre 1 y 2 (ambos inclusive).¿Cuántos productos tenemos?

df_baratos = df[df["price"].between(1,2, inclusive = "both")] 
df_baratos.shape
```
