```python
# 5.  ¿Cómo se distribuyen los precios de los productos en función de si se utilizan anuncios promocionales (`uses_ad_boosts`) o no? Vis I

# MEJOR HACERLO CON SEABORN, PORQUE MATPLOTLIB NO TIENE EL ARGUMENTO HUE
sns.histplot(x = "price", hue = "uses_ad_boosts", data = df, multiple='stack')
plt.ylabel("Conteo")
plt.xlabel("Precio");
```

```python
df['uses_ad_boosts'].value_counts()
```

```python
# Si lo quisieramos hacer con Matplotlib tendriamos que filtrar los datos en dos categorias, 
# con anuncios, y sin anuncios y superponer los dos histogramas, pero mejor hacerlo con seaborn. 

df_anuncios = df[df['uses_ad_boosts'] == 'Si']
df_sin_anuncios = df[df['uses_ad_boosts'] == 'No']

# Crear el histograma
plt.hist(df_anuncios['price'], alpha=0.5, label='Con anuncios', bins=30, color='orange')
plt.hist(df_sin_anuncios['price'], alpha=0.5, label='Sin anuncios', bins=30, color='blue')

# Etiquetas y leyenda
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.title('Distribución de Precios con y sin Anuncios')
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
# 6. ¿Existe una relación entre el uso de anuncios y las unidades vendidas? VIs I

# MEJOR CON SEABORN QUE TIENE EL ARGUMENTO HUE

sns.histplot(x = "mean_units_sold_per_product", hue = "uses_ad_boosts", data = df, bins = 20)
plt.ylabel("Conteo")
plt.xlabel("mean_units_sold_per_product");
```

```python
# 7. ¿Cuál es la distribución de las calificaciones de los empleados (`rating_employee`) para los productos que tienen el distintivo "Buena calidad del producto" (`badge_product_quality`) 
# y para los que no lo tienen? Haced que las dos gráficas tengan la misma escala en los dos ejes.

buenos = df[df["badge_product_quality"] == "Buena"]
excelente = df[df["badge_product_quality"] == "Excelente"]

excelente.head(2)
```

```python
# USANDO SUBPLOTS, PODEMOS PONER LOS DOS GRAFICOS UNO ENCIMA DEL OTRO PARA PODER COMPARARLOS BIEN, Y HACER LOS DOS HISTOGRAMAS POR SEPARADO
# CON SEABORN
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (20, 5), sharex= True)
sns.histplot(x = "rating_employee", data = buenos, ax = axes[0])
sns.histplot(x = "rating_employee", data = excelente, ax = axes[1])
axes[1].set_ylim([0, 150]);
```

```python
# USANDO SUBPLOTS, PODEMOS PONER LOS DOS GRAFICOS UNO ENCIMA DEL OTRO PARA PODER COMPARARLOS BIEN, Y HACER LOS DOS HISTOGRAMAS POR SEPARADO
# CON MATPLOTPLIB
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (20, 5), sharex= True)
axes[0].hist(x = "rating_employee", data = buenos, bins=60, edgecolor = 'black')
axes[1].hist(x = "rating_employee", data = excelente, bins=9, edgecolor = 'black')
axes[1].set_ylim([0, 150]);
```

```python
# SI LO COMBINAMOS EN UN GRAFICO SOLO USANDO BADGE_PRODUCT_QUALITY EN EL HUE DE SEABORN
plt.figure(figsize=(20, 2.5))
sns.histplot(x = "rating_employee", hue='badge_product_quality', data=df)
plt.ylim(0,150)
```

```python
# 9. ¿Cómo varía el precio promedio de los productos por país de origen? Muestra un gráfico de barras.VIs II
df_paises = df.groupby("origin_country")["price"].mean().reset_index().sort_values(by = "price", ascending = False)
df_paises
```

```python
sns.barplot(x = "origin_country", y = "price", data = df_paises, palette="viridis")
plt.ylabel("precio")
plt.xlabel("pais");
```

```python
# 11. Compara diferentes distribuciones de precios de productos en función de si tienen el distintivo "Producto local" o "Buena calidad del producto". Vis II


# Filtrar productos con distintivo "Producto local"
productos_local = df[df['badge_local_product'] == 'Si']

# Filtrar productos con distintivo "Buena calidad del producto"
productos_calidad = df[df['badge_product_quality'] == 'Buena']

# Crear subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Histograma para productos "Producto local"
axes[0].hist(productos_local['price'], bins=20, color='skyblue', alpha=0.7)
axes[0].set_title('Distribución de Precios (Producto Local)')
axes[0].set_xlabel('Precio')
axes[0].set_ylabel('Frecuencia')

# Histograma para productos "Buena calidad del producto"
axes[1].hist(productos_calidad['price'], bins=20, color='lightcoral', alpha=0.7)
axes[1].set_title('Distribución de Precios (Buena Calidad)')
axes[1].set_xlabel('Precio')
axes[1].set_ylabel('Frecuencia')

# Añadir título general
plt.suptitle('Comparación de Distribuciones de Precios')
plt.tight_layout()
plt.show()
```
